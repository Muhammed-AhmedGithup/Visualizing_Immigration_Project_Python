```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import pandas as pd
```

```python
plt.style.available
```

```python
mpl.style.use(['ggplot'])
```

```python

re =rq.get('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')

with open('Canada.csv','wb') as f:
    f.write(re.content)
    
df = pd.read_csv('Canada.csv')
df.head(5)
```

```python
df.set_index('Country',inplace=True)
```

```python
df.head(5)
```

```python
years=list(map(str,range(1980, 2014)))
```

```python
df.sort_values('Total',ascending=False,axis=0,inplace=True)
df5=df.head()
dftop5=df5[years].transpose()
dftop5.head()
```

```python
dftop5.plot(kind='area',stacked=False,figsize=(20,10),alpha=0.5)
plt.xlabel('Years')
plt.ylabel('immagrants')
```

```python
import numpy as np
count,bin=np.histogram(df['2013'])
print(count)
print(bin)
```

```python
df['2013'].plot(kind='hist',figsize=(8,5),xticks=bin)
plt.ylabel('number of country')
plt.xlabel('number of immagrants')
```

```python
# let's quickly view the dataset 
df.loc[['Denmark', 'Norway', 'Sweden'], years].plot(kind='hist')
```

```python
# let's quickly view the dataset 
df_t=df.loc[['Denmark', 'Norway', 'Sweden'], years].transpose()
df_t.head()
```

```python
count, binc=np.histogram(df_t, 15)
df_t.plot(kind='hist',figsize=(8,5),bins=15,xticks=binc,alpha=0.6,stacked=True)
plt.title('Histogram of Immigration from Denmark, Norway, and Sweden from 1980 - 2013')
plt.ylabel('Number of Years')
plt.xlabel('Number of Immigrants')
plt.show()
```

```python
df_ic=df.loc['Iceland',years]
df_ic.head()
```

```python
df_ic.plot(kind='bar',figsize=(8,5))
plt.xlabel('Year') # add to x-label to the plot
plt.ylabel('Number of immigrants') # add y-label to the plot
plt.title('Icelandic immigrants to Canada from 1980 to 2013') # add title to the plot
```

```python
df_con=df.groupby('Continent',axis=0).sum()
```

```python
df_con['Total'].plot(kind='pie',figsize=(10,5),startangle=90,shadow=True,autopct='%1.0f%%'
                    , labels=None, pctdistance=1.12)
plt.legend(labels=df_con.index,loc='upper left')
```

```python
df_japan = df.loc[['Japan'], years].transpose()
df_japan.head()
```

```python
df_japan.plot(kind='box',figsize=(8,5))
plt.ylabel('Number of immagrants')
```

```python
df_tot=pd.DataFrame(df[years].sum())
df_tot.reset_index(inplace=True)
df_tot.index=map(int,df_tot.index)
df_tot.columns=['Year','Total']
df_tot['Year']=df_tot['Year'].astype(float)
df_tot['Total']=df_tot['Total'].astype(float)
df_tot.head()
```

```python
df_tot.plot(kind='scatter',x='Year',y='Total',
            figsize=(8,5),color='r')

plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')

plt.show()
```

```python
x=df_tot['Year']
y=df_tot['Total']
fit=np.polyfit( x, y, deg=1)
fit
```

```python
df_tot.plot(kind='scatter',x='Year',y='Total',
            figsize=(8,5),color='r')
plt.title('Total Immigration to Canada from 1980 - 2013')
plt.xlabel('Year')
plt.ylabel('Number of Immigrants')


plt.plot(x,x*fit[0]+fit[1])
plt.annotate(f'y= {int(fit[0])}*x +{int(fit[1])}',xy=(2000,150000))

plt.show()
```

```python
years=list(map(str,range(1980,2014)))
```

```python
df_l=df[years]
imm_years=df_l.sum()
```

```python
fig, ax=plt.subplots()

imm_years.index=imm_years.index.map(int)
ax.plot(imm_years,
       marker='s',markersize=5,linestyle='dotted',color='g')

ax.set_title('Immigrants between 1980 to 2013') 
#Setting up the Labels
ax.set_xlabel('Years')
ax.set_ylabel('Total Immigrants')
plt.grid(True)
ax.legend(['Immagrants'])
plt.show()
```

```python
fig,ax=plt.subplots(figsize=(8,5))
imm_years.index=imm_years.index.map(int)

ax.scatter(imm_years.index,imm_years,marker='o',color='g',s=20)
ax.set_title('Immagrants between 1980 to 2013')
ax.set_xlabel('years')
ax.set_ylabel('Total immagrants')
ax.grid(True)
plt.show()
```

```python
df_=df.groupby(df.index)[years].sum()
df_n=df_.loc[['Denmark', 'Norway', 'Sweden'],years]
df_ns=df_n.T
```

```python
fig,ax=plt.subplots(figsize=(8,5))
ax.hist(df_ns)
ax.set_title('Immigration from Denmark, Norway, and Sweden from 1980 - 2013') 
ax.set_xlabel('Number of Immigrants')
ax.set_ylabel('Number of Years')
ax.legend(['Denmark', 'Norway', 'Sweden'])
#Display the plot
plt.show()
```

```python
fig ,ax=plt.subplots(figsize=(8,5))
ax.pie(imm_years[0:5],labels=years[0:5]
       ,autopct='%1.1f%%',explode=[0,0,0,0,0.1],
      colors=['gold','blue','lightgreen','coral','cyan'])

plt.title('Distribution of Immigrants from 1980 to 1985')

plt.show()
```

```python
fig ,ax=plt.subplots(1,2,sharey=True)
ax[0].plot(imm_years)
ax[0].set_title('Line plot on immigrants')

ax[1].scatter(imm_years.index,imm_years)
ax[1].set_title('Scatter plot on immigrants')
ax[0].set_ylabel('Number of immagrants')

fig.subtitle('Subplotting Example', fontsize=15)
fig.tight_layout()
plt.show
```

```python
fig=plt.figure(figsize=(8,5))
ax1=fig.add_subplot(1,2,1)
ax1.plot(imm_years)
ax1.set_title('Line plot on immigrants')

ax2=fig.add_subplot(1,2,2)
ax2.scatter(imm_years.index,imm_years)
ax2.set_title('Scatter plot on immigrants')

fig.subtitle('Subplotting Example',fontsize=15)
fig.light_layout()
plt.show()
```
