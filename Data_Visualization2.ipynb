```python
df_inc = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Police_Department_Incidents_-_Previous_Year__2016_.csv')

df_inc
```

```python
limit=100
df_inc=df_inc.loc[0:limit,:]
df_inc.head()
```

```python
latitude = 37.77
longitude = -122.42
```

```python
map_san=fl.Map(location=[latitude,longitude],zoom_start=12)
map_san
```

incs=fl.map.FeatureGroup()

```python
incs=fl.map.FeatureGroup()
```

```python
for lat,lng in zip(df_inc.Y,df_inc.X):
    incs.add_child(
        fl.vector_layers.CircleMarker(
            [lat,lng],
            color='yellow',
            radius=5,
            fill=True,
            fill_color='blue',
            fill_opacity=0.6
           
        )
    )
    
    
```

```python
map_san.add_child(incs)
```

```python

for lat,lng,label in zip(df_inc.Y,df_inc.X,df_inc.Category):
    fl.Marker([lat,lng]).add_to(map_san)
```

```python
map_san
```

```python
mapsan=fl.Map(location=[latitude, longitude],zoom_start=12)

for lat,lng in zip(df_inc.Y,df_inc.X):
    fl.vector_layers.CircleMarker(
        [lat,lng],
        color='yellow',
        radius=5,
        fill_color='blue',
        fill_opacity=0.6,
        fill=True
    
    ).add_to(mapsan)
    
mapsan
```

```python
from folium import plugins

sanMap=fl.Map(location=[latitude, longitude],zoom_start=12)
incss=plugins.MarkerCluster().add_to(sanMap)
for lat , lng in zip(df_inc.Y,df_inc.X):
    fl.Marker(
        [lat,lng]
    ).add_to(incss)
sanMap
```

```python
df_can = pd.read_csv('https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DV0101EN-SkillsNetwork/Data%20Files/Canada.csv')
df_can.head()
```
